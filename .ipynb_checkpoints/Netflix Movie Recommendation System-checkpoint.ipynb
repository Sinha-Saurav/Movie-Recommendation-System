```python
def fetch_genres(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])

    return l
```

```python
fetch_genres(df['genres'][0])
```

```python
df['genres'] = df['genres'].apply(fetch_genres)
```

```python
df[:3]
```

## Keywords

```python
df['keywords'][0]
```

```python
def fetch_keywords(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])

    return l
```

```python
fetch_keywords(df['keywords'][0])
```

```python
df['keywords'] = df['keywords'].apply(fetch_keywords)
```

```python
df[:3]
```

```python
df['cast'][0]
```

```python
def fetch_cast(text):
    l = []
    count = 0;
    for i in ast.literal_eval(text):
        if count<3:
            l.append(i['name'])
            count+=1
        else:
            break

    return l
```

```python
fetch_cast(df['cast'][0])
```

```python
df['cast'] = df['cast'].apply(fetch_cast)
```

```python
df[:3]
```

# Crew

```python
df['crew'][0]
```

```python
ast.literal_eval(df['crew'][0])
```

```python
def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])

    return l
```

```python
fetch_director(df['crew'][0])
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

# NLP CONCEPTS TO PREPROCESS THE TEST DATA

- lower case
- tokenization
- stemming
- stopwords removal
- Encode the Data

```python
ps = PorterStemmer()

def text_preprocess(text):

    clean = []
    for i in text.split():
        lower = i.lower()
        clean.append(ps.stem(lower))
    return " ".join(clean)    

text_preprocess(data['tags'][0])
```

```python
data['tags'].apply(text_preprocess)
```

```python
data.head()
```

# final function to recommend top 7 similar movies

```python
def recommend(movie):
    movie_index = data[data['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list= sorted(list(enumerate(distance)), reverse = True, key = lambda x:x[1])[1:9]

    for i in movie_list:
        print(data.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```

```python
recommend('Harry Potter and the Chamber of Secrets')
```
